```python
import pandas as pd
import datetime
!pip install datetime_truncate
from datetime_truncate import truncate
import numpy as np
import math
```

```python
data = pd.read_csv("../data/Suicide_withDBPedia_Values_fr.csv", sep=",", encoding='utf-8-sig')
data.head()
```

## List all values in birthCountry to know which ones need to be changed

```python
birthCountry = data["birthCountry"]
birthCountry.unique()
```

## Replacing values

```python
for i in range(len(data)):
    if (birthCountry[i]=="United_States_of_America" or birthCountry[i] == "Essex" or birthCountry[i] == "Ulster" or birthCountry[i] == "Suffolk" or birthCountry[i] == "United_States"):
        birthCountry[i] = "USA"
    elif(birthCountry[i] == "Kent" or birthCountry[i] == "United_Kingdom"):
        birthCountry[i] = "UK"
    elif(birthCountry[i] == "Georgia_(country)"):
        birthCountry[i] = "Georgia"
    elif(birthCountry[i] == "Republic_of_Ireland" or birthCountry[i] == "Northern_Ireland"):
        birthCountry[i] = "Ireland"
```

```python
data["birthCountry"] = birthCountry
```

## Deleting the ones for who the deathdate is incorrect

For those ones, it is because they were dead before the 17th century. 

```python
data = data[data["Name"] != "Xiang_Yu"]
data = data[data["Name"] != "Quintillus"]
data = data[data["Name"] != "Wei_Zhongxian"]
data = data[data["Name"] != "Francesco_Borromini"]
data = data[data["Name"] != "Qin_Er_Shi"]
data = data[data["Name"] != "Bartolomeo_Schedoni"]
data = data[data["Name"] != "Uriel_da_Costa"]
data = data[data["Name"] != "Joseph_Newton_Chandler_III"]
data = data[data["Name"] != "Qu_Yuan"]
data = data[data["Name"] != "Chimalpopoca"]
data = data[data["Name"] != "François_Vatel"]
data = data[data["Name"] != "Hōjō_Ujimasa"]
data = data[data["Name"] != "Oda_Nobunaga"]
data = data[data["Name"] != "Oda_Nobutada"]
data = data[data["Name"] != "Ōuchi_Yoshitaka"]
data = data[data["Name"] != "Sen_no_Rikyū"]
data = data[data["Name"] != "Shibata_Katsuie"]
```

```python

dateDeathTimestamp = []
for i in range(len(data)):
    dateDeathTimestamp.append(pd.Timestamp(data["deathdate"].values[i])) #convert into timestamp
    dateDeathTimestamp[i] = truncate(dateDeathTimestamp[i], 'year') # trunc to years
```

```python
deathDateCountry = pd.DataFrame()
deathDateCountry["deathdate"] = dateDeathTimestamp
Ones = np.ones(len(deathDateCountry))
deathDateCountry["Country"] = data["birthCountry"]
deathDateCountry["Ones"] = Ones
deathDateCountry = deathDateCountry.dropna()
deathDateCountry.head()
```

```python
grouped_deathDateCountry = deathDateCountry.groupby(["Country", "deathdate"]).sum().reset_index()
```

```python
grouped_deathDateCountry.head()
```

```python
nb = [0]*len(grouped_deathDateCountry)
current_nb = 1
for i in range(len(grouped_deathDateCountry)):
    if (i > 0):
        if(grouped_deathDateCountry["Country"].values[i] == grouped_deathDateCountry["Country"].values[i-1]):
            current_nb += grouped_deathDateCountry["Ones"].values[i]
            nb[i] = current_nb
        else:
            current_nb = grouped_deathDateCountry["Ones"].values[i]
```

```python
grouped_deathDateCountry["Count"] = nb
grouped_deathDateCountry.head()
```

```python
data_pivot = grouped_deathDateCountry.pivot_table("Count", "Country", "deathdate")
data_pivot = data_pivot.fillna(0)
data_pivot.head()
```

```python
data_pivot = data_pivot.iloc[:, 100:-2]
data_pivot = data_pivot.cumsum(axis=1)
total = data_pivot.sum(axis=0).T
data_pivot = (data_pivot / total)*100
data_pivot = data_pivot.reset_index()
data_pivot
```

```python
data_pivot.to_csv('../data/Dataframe_for_racing_plot.csv', sep=',', encoding='utf-8-sig', index=False)
```
