```python
# NLTK Stop words
from nltk.corpus import stopwords
stop_words = stopwords.words('french')
stop_words.extend(['enfin', 'modifier', 'jusqu','part','peu','them','souvent','dedie','penser','presenton','eter','acc','esser','ainsi','non','partir','oeuvre','si','ete','cet', 'article', 'faire', 'bon', 'permettre', 'permet', 'proposer', 'proposons', 'obtenir', 'obtenons','tel', 'présenter', 'présentons', 'quel','face','mettre','basee','issu','gerer','but','etre','source','tre','pre','lie','sous', 'forme', 'faire', 'être', 'ete', 'donner', 'partie', 'étape', 'utiliser', 'utilisons', 'être', 'plus', 'fin', 'contexte'])
```

```python
data = pd.read_csv("../data/Suicides_wikipedia_scrapping_fr.csv", encoding="utf-8-sig", delimiter=",")
data.head()
```

#### get all the names of our biographies and add them to stop_words

```python
Names = data["Name"]
all_names = ""

for name in range (len(Names)):
    Names[name] = Names[name].replace("_", " ")
    Names[name] = Names[name].replace("(", "")
    Names[name] = Names[name].replace(")", "")
    all_names += " " + Names[name].lower()
all_names = all_names.split() 
print(all_names)
```

```python
stop_words.extend(all_names)
```

```python
Biography = data.Content.values.tolist()
# Remove Emails
Biography = [re.sub('\S*@\S*\s?', '', sent) for sent in Biography]

# Remove new line characters
Biography = [re.sub('\s+', '', sent) for sent in Biography]

# Remove distracting single quotes
Biography = [re.sub("\'", "", sent) for sent in Biography]

# Remove the commas
Biography = [re.sub(",", " ", sent) for sent in Biography]
```

```python
Biography[1]
```

```python
Biography_clean_brackets = ['']*len(Biography)
for i in range(len(Biography)):
    for j in range(69, len(Biography[i]) - 7):
        Biography_clean_brackets[i]+=Biography[i][j]
```

```python
Biography_clean_brackets[2]
```

```python
def sent_to_words(sentences):
    for sentence in sentences:
        yield(gensim.utils.simple_preprocess(str(sentence), deacc=True))  # deacc=True removes punctuations

data_words = list(sent_to_words(Biography_clean_brackets))
```

```python
data_words[2]
```

```python
# Build the bigram and trigram models
bigram = gensim.models.Phrases(data_words, min_count=5, threshold=100) # higher threshold fewer phrases.
trigram = gensim.models.Phrases(bigram[data_words], threshold=100)  

# Faster way to get a sentence clubbed as a trigram/bigram
bigram_mod = gensim.models.phrases.Phraser(bigram)
trigram_mod = gensim.models.phrases.Phraser(trigram)

# See trigram example
print(trigram_mod[bigram_mod[data_words[2]]])
```

```python
# Define functions for stopwords, bigrams, trigrams and lemmatization
def remove_stopwords(texts):
    return [[word for word in simple_preprocess(str(doc)) if word not in stop_words] for doc in texts]

def make_bigrams(texts):
    return [bigram_mod[doc] for doc in texts]

def make_trigrams(texts):
    return [trigram_mod[bigram_mod[doc]] for doc in texts]

def lemmatization(texts, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV']):
    """https://spacy.io/api/annotation"""
    texts_out = []
    for sent in texts:
        doc = nlp(" ".join(sent)) 
        texts_out.append([token.lemma_ for token in doc if token.pos_ in allowed_postags])
    return texts_out
```

```python
# Remove Stop Words
data_words_nostops = remove_stopwords(data_words)

# Form Bigrams
data_words_bigrams = make_bigrams(data_words_nostops)

# Initialize spacy 'en' model, keeping only tagger component (for efficiency)
# python3 -m spacy download en
nlp = spacy.load('fr', disable=['parser', 'ner'])

# Do lemmatization keeping only noun, adj, vb, adv
data_lemmatized = lemmatization(data_words_bigrams, allowed_postags=['NOUN', 'ADJ', 'VERB', 'ADV'])

#print(stop_words)
data_lemmatized = remove_stopwords(data_lemmatized)

print(data_lemmatized[:2])
```

```python
import pandas as pd 
from nltk.tokenize.treebank import TreebankWordDetokenizer

for biography in range(len(data_lemmatized)):
    data_lemmatized[biography] = TreebankWordDetokenizer().detokenize(data_lemmatized[biography])
new_dataframe = pd.DataFrame()
new_dataframe["biography_cleaned"] = data_lemmatized
new_dataframe["name"] = data["Name"]
new_dataframe.head()
```
